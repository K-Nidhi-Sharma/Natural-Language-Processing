### Import libraries

```python
import wikipedia
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
```

### Extracting data from wikipedia

```python
articles=['Data Science','Artificial intelligence','Data Analysis','Data Analytics','Deep Learning']
wiki_lst=[]
title=[]
for article in articles:
   print("loading content: ",article)
   wiki_lst.append(wikipedia.page(article).content)
   title.append(article)
print("examine content")
wiki_lst
```

```python
print("Articles in the data: ")
title
```

### Number of articles in data 

```python
print("The number of articles in the data are: ",len(wiki_lst))
```

### TF-IDF Vectorization of the articles

```python
tfidf=TfidfVectorizer()
train_tfidf = tfidf.fit_transform(wiki_lst)
train_tfidf
```

### Query 

```python
q=['data science is an upcoming field',"Time series helps in forecasting","Artificial intelligence is the future helping in automation and uses brain mechanism","Natural language processig helps in sentiment analysis", "Data analysis helps in anlaysing the data and find conclusions"]
```

### Vectorizing query

```python
query_vec1 = tfidf.transform([q[0]])
query_vec2 = tfidf.transform([q[1]])
query_vec3 = tfidf.transform([q[2]])
query_vec4 = tfidf.transform([q[3]])
query_vec5 = tfidf.transform([q[4]])
```
